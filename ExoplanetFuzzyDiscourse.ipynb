```python
import numpy as np
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import skfuzzy as fuzz
from skfuzzy import control as ctrl
# Define the universe for each variable
x_depth = np.linspace(0, 3000, 500)
x_duration = np.linspace(0, 6, 500)
x_interval = np.linspace(0, 6, 500)
x_likelihood = np.linspace(0, 1, 500)

# Define the membership functions for each variable
depth_small = np.maximum(np.minimum((x_depth-0)/(500-0), (1000-x_depth)/(1000-500)), 0)
depth_medium = np.maximum(np.minimum((x_depth-500)/(1500-500), (2000-x_depth)/(2000-1500)), 0)
depth_large = np.maximum(np.minimum((x_depth-1500)/(2500-1500), (3000-x_depth)/(3000-2500)), 0)

duration_short = np.maximum(np.minimum((x_duration-0)/(1-0), (2-x_duration)/(2-1)), 0)
duration_medium = np.maximum(np.minimum((x_duration-1.5)/(3-1.5), (4.5-x_duration)/(4.5-3)), 0)
duration_long = np.maximum(np.minimum((x_duration-4)/(5-4), (6-x_duration)/(6-5)), 0)

interval_short = np.maximum(np.minimum((x_interval-0)/(1-0), (2-x_interval)/(2-1)), 0)
interval_medium = np.maximum(np.minimum((x_interval-1.5)/(3-1.5), (4.5-x_interval)/(4.5-3)), 0)
interval_long = np.maximum(np.minimum((x_interval-4)/(5-4), (6-x_interval)/(6-5)), 0)

likelihood_unlikely = np.maximum(np.minimum((x_likelihood - 0) / (0.25 - 0), (0.75 - x_likelihood) / (0.75 - 0.25)), 0)
likelihood_likely = np.maximum(np.minimum((x_likelihood - 0.25) / (0.75 - 0.25), (1 - x_likelihood) / (1 - 0.75)), 0)

# Define the given likelihood values
likelihood_values = [0.00, 0.33]

# Plot the membership functions
fig, axs = plt.subplots(4, figsize=(8,10))

axs[0].plot(x_depth ,depth_small ,'b' ,linewidth=1.5 ,label='Small')
axs[0].plot(x_depth ,depth_medium ,'g' ,linewidth=1.5 ,label='Medium')
axs[0].plot(x_depth ,depth_large ,'r' ,linewidth=1.5 ,label='Large')
axs[0].set_title('Depth of Dip')
axs[0].legend()

axs[1].plot(x_duration ,duration_short ,'b' ,linewidth=1.5 ,label='Short')
axs[1].plot(x_duration ,duration_medium ,'g' ,linewidth=1.5 ,label='Medium')
axs[1].plot(x_duration ,duration_long ,'r' ,linewidth=1.5 ,label='Long')
axs[1].set_title('Duration of Dip')
axs[1].legend()

axs[2].plot(x_interval ,interval_short ,'b' ,linewidth=1.5 ,label='Short')
axs[2].plot(x_interval ,interval_medium ,'g' ,linewidth=1.5 ,label='Medium')
axs[2].plot(x_interval ,interval_long ,'r' ,linewidth=1.5 ,label='Long')
axs[2].set_title('Interval Between Dips')
axs[2].legend()

axs[3].plot(x_likelihood ,likelihood_unlikely ,'b' ,linewidth=1.5 ,label='Unlikely')
axs[3].plot(x_likelihood ,likelihood_likely ,'g' ,linewidth=1.5 ,label='Likely')
axs[3].set_title('Exoplanet Likelihood')
axs[3].legend()

#


axs[3].set_title('Exoplanet Likelihood')
axs[3].legend()

plt.tight_layout()
plt.show()
```
